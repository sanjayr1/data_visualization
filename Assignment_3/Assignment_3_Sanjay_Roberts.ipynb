```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

Assignment 3, Part 1: Using the built-in Seaborn dataset mpg, provide the heatmap of the correlation of all the numeric columns and provide the pairplot of the same.

```python
df = sns.load_dataset('mpg')
df.head()
```

```python
sns.heatmap(df.corr(), cmap='coolwarm')
plt.tight_layout()
plt.show()
```

```python
sns.pairplot(df.corr()).map(plt.scatter)
plt.tight_layout()
plt.show()
```

Assignment 3, Part 2: Using the built-in Seaborn dataset diamonds, establish a FacetGrid based on ‘cut’ and ‘color’. Eliminate colors ‘D’ and ‘E’ as well as the cut ‘Fair’. Within that grid, plot the scatterplot for ‘price’ vs. ‘carat’.


```python
df = sns.load_dataset('diamonds')
df.head()
```

```python
df = df[df.color != 'D']
df = df[df.color != 'E']
df = df[df.cut != 'Fair']
```

```python
g2 = sns.FacetGrid(df, col='cut', row='color')
g2 = g2.map(plt.scatter,'price',
  'carat')
plt.show(g2)
```

Assignment 3, Part 3: Using the built-in Seaborn dataset car_crashes, prepare plots with the scatter gram with the linear model for both the total vs. speeding and the total vs. alcohol.

```python
df = sns.load_dataset('car_crashes')
df.head()
```

```python
sns.lmplot(data = df, x='total',
  y='speeding', fit_reg=True)
plt.show()

sns.lmplot(data=df, x='total',
  y='alcohol',
  fit_reg= True)
plt.show()
```

Assignment 4, Part 4: Using the built-in Seaborn dataset iris, provide a plot with four subplots wherein the distribution of each of the numeric columns is presented as a set of boxplots, one for each ‘species’.

```python
df = sns.load_dataset('iris')
df.head()
```

```python
fig,axes=plt.subplots(2,2,figsize=(16,9))
sns.boxplot(x = df.sepal_length, y=df.species, data = df,ax=axes[0,0])
sns.boxplot(x = df.sepal_width, y=df.species, data = df, ax=axes[1,0])
sns.boxplot(x = df.petal_length, y=df.species, data = df, ax=axes[0,1])
sns.boxplot(x = df.petal_width, y=df.species, data = df, ax=axes[1,1])
plt.tight_layout()
plt.show()
```
